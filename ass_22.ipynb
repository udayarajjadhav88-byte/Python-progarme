```python
store_brand="FreshMart Express"
customer_id="USER_9921"
shipping_fee=5.99

basket=[
    ("Oranganic Bananas",3.50),
    ("Whole milk",4.25),
    ("Sourdough Bread",5.00),
    ("Avocados",6.50)
]
basket.append(("Chocate Bar",2.25))

item_price=[item[1] for item in basket]
subtotal=sum(item_price)
discount=3.00 

total_tax=subtotal*0.05
grand_total=(subtotal-discount)+total_tax+shipping_fee
print(f"--{store_brand}RECEIPT--")
print(f"Customer ID:{customer_id}")
print("-"*30)

for name,price in basket:
    print(f"{name:<0}${price:>602f}")
    print("-" * 30)
print(f"Basket Subtotal:    ${subtotal:>6.2f}")
print(f"Bulk Discount:     -${discount:>6.2f}")
print(f"Sales Tax (5%):     ${total_tax:>6.2f}")
print(f"Shipping Fee:       ${shipping_fee:>6.2f}")
print("=" * 30)
print(f"GRAND TOTAL:        ${grand_total:>6.2f}")
print("=" * 30)
```
