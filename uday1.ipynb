```python
print("Available Courses:")
print("1. Python Programming")
print("2. Data Analytics")
print("3. AI & ML")

course_choice = int(input("Enter course number (1-3): "))
student_discount = input("Are you a student? (yes/no): ")
early_discount = input("Early registration? (yes/no): ")

if course_choice == 1:
    course_name = "Python Programming"
    fee = 5000
elif course_choice == 2:
    course_name = "Data Analytics"
    fee = 8000
elif course_choice == 3:
    course_name = "AI & ML"
    fee = 12000
else:
    print("Invalid course selection")
    exit()

discount = 0

if student_discount.lower() == "yes":
    discount += fee * 0.10

if early_discount.lower() == "yes":
    discount += fee * 0.05

final_amount = fee - discount

print("\n----- Course Fee Summary -----")
print("Course Name          :", course_name)
print("Original Fee         : ₹", fee)
print("Total Discount       : ₹", discount)
print("Final Payable Amount : ₹", final_amount)
```
